```python
import pandas as pd
import seaborn as sns
sns.set(rc={'figure.figsize':(20.7,8.27),"font.size":30,"axes.titlesize":20,"axes.labelsize":20},style="white", context='paper')
import matplotlib.pyplot as plt
```

```python
logpmatrix = pd.read_csv('results_adj_nlog10_allVar', sep = '\t', header=None, low_memory=False)
logpmatrix.columns =['SNP', 'pos']+ ['BIO'+ str(i) for i in range(1,21)]
logpmatrix['index']= logpmatrix.index
```

```python
for i in ['BIO'+ str(i) for i in range(1,21)]:
    sns.scatterplot(x='pos', y=i, data= logpmatrix, marker='+')
    sns.despine()
    plt.ylim([0,9])
    plt.xlim([0, logpmatrix.shape[0]])
    plt.show()
```
